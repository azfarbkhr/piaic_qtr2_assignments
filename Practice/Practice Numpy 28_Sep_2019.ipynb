```python
count_supplier_pd = df['SupplierName'].value_counts()
count_supplier_pd[:10]
```

```python
grouped_df = df.groupby(['SupplierName','CategoryName'])
grouped_df['SupplierName','CategoryName'].head(5)
```

```python
group_count = grouped_df.size().unstack().fillna(0)
indexer = group_count.sum(1).argsort()
count_subset = group_count.take(indexer)[-10:]
count_subset
```
